#### Converting Datetime to date data type

#### Summarising games, deposit and withdrawal data for each user, date and slot

```python
total_games= df_game.groupby(['User ID',"Date_cln", "time_slot"],as_index=False).agg(tot_game= ('Games Played','count')).\
rename(columns={'User ID': 'User Id'})
total_games.head()

dep_amt= df_dep.groupby(['User Id',"Date_cln", "time_slot"],as_index=False).agg(tot_amt_dp=('Amount','sum'), tot_cnt_dp=('User Id', 'count'))
dep_amt.head(2)

wd_amt= df_wd.groupby(['User Id',"Date_cln", "time_slot"],as_index=False).agg(tot_amt_wd=('Amount','sum'),tot_cnt_wd=('User Id', 'count'))
wd_amt.head(2)
```

#### Joining games, deposit and withdrawal summarised data

```python
df_jn = total_games.merge(dep_amt ,on=['User Id',"Date_cln", "time_slot"], how = 'left').\
        merge(wd_amt, on=['User Id',"Date_cln", "time_slot"],how = 'left')
df_jn.head()
```

```python
#replacing Nulls by 0
df_jn.loc[:,['tot_amt_dp','tot_cnt_dp','tot_amt_wd','tot_cnt_wd']] = df_jn.loc[:,['tot_amt_dp','tot_cnt_dp','tot_amt_wd','tot_cnt_wd']].fillna(0)
df_jn.head()
#df_jn.to_csv('D:\\4_interview\\Vindiata\\joined_data_vindiata_v2.csv', index= False)
```

##### Loyalty Point = (0.01 * deposit amount) + (0.005 * Withdrawal amount) + (0.001 * (maximum of (#deposit - #withdrawal) or 0)) + (0.2 * Number of games played)

### My Observations 
##### Loyalty Point is calculated as a factor of 
1. **Deposit Amount** with weights as **1%**
2. **Withdrawal Amount** with weights as **0.5%**
3. **Count of Deposits and withdrawals** with weights as **0.1%**
4. **Number of games** plated with weights as **20%**

**Case 1 - Loyalty points alone** as a metric for distributing cash benefits to top 50 applicants as already during calculation of loyalty points, **20% weightage is given to Number of games played by user**
1. Take data of top 50 users basis loyalty points
2. Distribute Rs 50K in the ration of their loyalty points
3. prize = *50K x (loyalty point of a player)/ total loyalty points*

**Case 2 - Loyalty points & games played** using both as a metric, we can give 80% weightage to loyalty and 20% weightage to games played
1. prize = *50K x [0.8 x (loyalty point of a player)/ (total loyalty points) + 0.2 x (games played by a player)/ (total games played)]*

**Recommendation** : We can go ahead with **Case 2** scenario as the players who have played more number of games have **dedicated more time** to gaming and hence they should be rewarded more

```python
top_50 = df_jn_rnk.loc[df_jn_rnk["rank"] <=50,:]
top_50["prize_case1"] =  50000*top_50.loyalty_pt/top_50.loyalty_pt.sum()
top_50["prize_case2"] =  50000*(0.8*top_50.loyalty_pt/top_50.loyalty_pt.sum() + 0.2*top_50.tot_game/top_50.tot_game.sum())

top_50
```
